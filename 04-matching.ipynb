# Matching ILPs
## A first, bad formulation
Let us import the gurobi library and set up a first optimization model.

```python
import gurobipy as grb
model = grb.Model(name="Matching Bad Formulation")
```

We will try the following formulation:
  \begin{align*}
      \max &\sum_{e \in E} w(e)\\
      \text{ subject to } \quad & x_e + x_{e'} \le 1 && \text{ for all } e,e' \in E \text{ with } |e \cap e'| = 1 \\
      & x_e \in \{0,1\} && \text{ for all } e \in E
  \end{align*}

### Generate variables
We need one variable `x[i,j]` per edge i,j. We will use a gurobi `tupledict` to hold these variables (a dictionary with tuples as keys).

```python
x = grb.tupledict()
for i,j, weight in G.edges.data('weight'):
    x[i,j] = model.addVar(obj=weight, vtype=grb.GRB.BINARY)
```

### Constraints
Next, we add all constraints.

```python
for u,v in G.edges():
    for i,j in G.edges():
        if len({u,v}.intersection({i,j})) == 1:
            model.addConstr(x[u,v] + x[i,j] <= 1)
```

### Solve
Let gurobi solve the problem.

```python
model.ModelSense = grb.GRB.MAXIMIZE
model.optimize()
```

### Access results
The value of an edge variable in the optimal solution can be accessed via `x[i, j].x`.
Let us collect all edges in the optimal matching and plot them.

```python
matching = [(i,j) for i,j in G.edges if x[i,j].x > 0.9]
pos = [(pos['x'], pos['y']) for i,pos in G.nodes(data=True)]
nx.draw_networkx_nodes(G, pos, node_color='#66a3ff', node_size=500)
nx.draw_networkx_labels(G, pos, font_weight='bold' )
nx.draw_networkx_edges(G, pos,  edge_color='#e6e6e6')
nx.draw_networkx_edges(G, pos,  edgelist=matching, edge_color='#ffa31a', width=4)
```

## A better formulation
A much better formulation should be the following one.
\begin{align*}
    \max &\sum_{e \in E} w(e)\\
    \text{ subject to } \quad & \sum_{e \in \delta(v)} x_e \le 1 && \text{ for all } v \in V \\
    & x_e \in \{0,1\} && \text{ for all } e \in E
\end{align*}
Here, $ \delta(v) \subseteq E $ is the set of all edges incident to $ v $.

### Model formulation in Gurobi

```python
model = grb.Model(name="Matching Better Formulation")

x = grb.tupledict()
for i,j, weight in G.edges.data('weight'):
    x[i,j] = model.addVar(obj=weight, vtype=grb.GRB.BINARY)

for v in G.nodes():
    model.addConstr(sum([x[i,j] for i,j in x.keys() if (v==i) or (v==j)]) <= 1)
```

### Solution
Let's see how gurobi performs on this formulation.

```python
model.ModelSense = grb.GRB.MAXIMIZE
model.optimize()
```

Again, let's plot the computed matching.

```python
matching = [(i,j) for i,j in G.edges if x[i,j].x > 0.9]
pos = [(pos['x'], pos['y']) for i,pos in G.nodes(data=True)]
nx.draw_networkx_nodes(G, pos, node_color='#66a3ff', node_size=500)
nx.draw_networkx_labels(G, pos, font_weight='bold' )
nx.draw_networkx_edges(G, pos,  edge_color='#e6e6e6')
nx.draw_networkx_edges(G, pos,  edgelist=matching, edge_color='#ffa31a', width=4)
```
